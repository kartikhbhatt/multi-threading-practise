```python
import threading
import time
```

```python
def print_numbers():
    for i in range(5):
        time.sleep(1)
        print(f"number : {i}")
    return None
```

```python
def print_letter():
    for i in "abcd":
        time.sleep(1)
        print(f"letter : {i}")
    return None
```

```python
def main():
    # single threaded function waits for the dependencies to resolve then completes pending execution
    t = time.time()
    print_letter()
    print_numbers()
    finished_time = time.time() - t
    print(f"finished execution in for single threaded function: {finished_time}")


    # multithreaded process runs concurrently/simultaneously
    t=time.time()
    t1 = threading.Thread(target=print_letter)
    t2 = threading.Thread(target=print_numbers)
    # starting threads
    t1.start()
    t2.start()
    # wait for the threads to finish then calculate execution time
    t1.join()
    t2.join()

    finished_time = time.time() - t
    print(f"finished execution in for multi threaded function: {finished_time}")

```

```python
if __name__ == "__main__":
    main()
```
